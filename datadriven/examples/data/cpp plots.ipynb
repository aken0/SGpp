```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.integrate import nquad
%matplotlib inline
```

```python
ee=np.genfromtxt("adap2(level).txt",delimiter=',')
```

```python
plt.figure(figsize=(5,5),dpi=300)

plt.plot(ee[:,0],ee[:,1],".")
```

```python
def e(x):
    return (x**3-x**2)
ye=nquad(e,[[-1,1]])
ye2=nquad(lambda x:1+(np.sin(x)+np.cos(x))/np.exp(x),[[-1,1]])
```

```python
df=pd.read_csv('regular.txt',header=None)
plt.figure(figsize=(7,5),dpi=200)
plt.gca().set_yscale('log')
plt.xlabel('level')
plt.plot(range(1,24),abs(df.loc[1]-2),label='1.0')
plt.plot(range(1,24),abs(df.loc[0]-ye[0]),label='x^3-x^2')
plt.plot(range(1,24),abs(df.loc[2]-ye2[0]),label='1+(sin(x)+cos(x))/e^x')
plt.legend()
plt.savefig("regular.png")
plt.show()
```
